```python
#Basic initalization of linklist
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
class LinkedList:
    def __init__(self):
        self.head=None
        
    def printList(self):
        temp=self.head
        while (temp):
            print(temp.data,end="-->")
            temp=temp.next


if __name__=="__main__":
    llist=LinkedList()
    
    llist.head=Node(1)
    second=Node(2)
    third=Node(3)
    
    llist.head.next=second
    second.next=third
    
    llist.printList()
```

```python
#Basic traversal of linklist
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
class LinkedList:
    def __init__(self):
        self.head=None
    
    
    def printList(self):
        temp=self.head
        while (temp):
            print(temp.data,end=" ")
            temp=temp.next

if __name__=="__main__":
    llist=LinkedList()
    
    llist.head=Node(1)
    second=Node(2)
    third=Node(3)
    
    llist.head.next=second
    second.next=third
    
    llist.printList()
```

```python
#Alternate approach with using deque pre-defined function 
import collections as co
#llist=co.deque("hello" ,maxlen=5) # maxlen is use for deque len set's a limit.
llist=co.deque()
llist.append(1)
llist.append(2)
llist.append(3)
llist.appendleft(0)

print("Element in the linked list : ",llist)

llist.insert(1,5)
print(llist)

print(llist.pop())
print(llist.popleft())
print(llist)

llist.remove(1)
print(llist)


llist.clear()
print(llist)

llist.extend("123")
llist.extend("Tatha")
llist.extendleft("hey")
print(llist)

llist.rotate(-1)
print(llist)
```

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.ref=None

class Linked_list:
    def __init__(self):
        self.head=None
    
    def show_LL(self):
        if self.head is None:
            print("LL is empty")
        else:
            n=self.head
            while n is not None:
                print(n.data,"-->",end=" ")
                n=n.ref
                
    def add_begin(self,data):
        new_node=Node(data)
        new_node.ref=self.head
        self.head=new_node
        
    def add_end(self,data):
        new_node=Node(data)
        if self.head==None:
            self.head=new_node
        else:
            n=self.head
            while n.ref is not None:
                n=n.ref
            n.ref=new_node
            
    def add_after_node(self,data,x):
        n=self.head
        while n is not None :
            if x==n.data:
                break
            n=n.ref
        if n is None:
            print("data is not present in LL")
        else:
            new_node = Node(data)
            new_node.ref=n.ref
            n.ref=new_node
    
    def add_before_node(self,data,x):
        if self.head is None:
            print("LL is empty")
            return 
        if self.head.data == x:
            new_node=Node(data)
            new_node.ref=self.head
            self.head=new_node
            return
        n=self.head
        while n is not None:
            if n.ref.data==x:
                break
            n=n.ref
        if n.ref is None :
            print("Node is not found")
        else:
            new_node=Node(data)
            new_node.ref=n.ref
            n.ref=new_node
            
#     def insert_empty(self,data):
#         if self.head==None:
#             new_node=Node(data)
#             self.head=new_node
#         else:
#             print("LL is not empty!")
            
    
    def del_begin(self):
        if self.head is None:
            print("LL is empty!")
        else:
            self.head=self.head.ref
    
    def del_end(self):
        if self.head is None:
            print("LL is empty")
        else:
            n=self.head
            while n.ref.ref is not None:
                n=n.ref
            n.ref = None
    
    def del_mid(self,x):
        n=self.head
        if self.head==None:
            print("LL is empty")
        if self.head.data == x:
            self.head = self.head.ref
            return 
        else:
            while n.ref is not None:
                if n.ref.data == x:
                    break
                n=n.ref
            if n.ref is None:
                print("data does not found")
            else:
                n.ref=n.ref.ref
                        
LL=Linked_list()

LL.add_begin(10)
LL.add_begin(20)
LL.add_begin(30)
LL.add_begin(40)
#LL.del_end()

LL.show_LL()
```

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.nref=None
        self.pref=None
        
class doubleLL:
    def __init__(self):
        self.head=None
        
    def forward_LL(self):
        if self.head is None :
            print("LL is empty")
        else:
            n=self.head
            while n is not None:
                print(n.data,"-->",end=" ")
                n=n.nref
                
    def backward_LL(self):
        if self.head==None:
            print("LL is empty")
        else:
            n=self.head
            while n.nref is not None:
                n=n.nref
            while n is not None:
                print(n.data,"-->",end=" ")
                n=n.pref
                
    def insert_empty(self,data):
        if self.head == None:
            new_node=Node(data)
            self.head=new_node
        else:
            print("DLL is not empty")
            
    def add_begin(self,data):
        new_node=Node(data)
        if self.head==None:
            self.head=new_node
        else:
            new_node.nref=self.head
            self.head.pref=new_node
            self.head=new_node
    
    def add_end(self,data):
        new_node=Node(data)
        if self.head==None:
            new_head=new_node
        else:
            n=self.head
            while n.nref is not None:
                n=n.nref
            n.nref=new_node
            new_node.pref=n
    
    def add_after(self,data,x):
        
        if self.head==None:
            print("LL is empty")
        else:
            n=self.head
            while n is not None:
                if n.data==x:
                    break
                n=n.nref
            if n is None:
                print("x not in DLL")
            else:
                new_node=Node(data)
                new_node.nref=n.nref
                new_node.pref=n
                if n.nref is not None:
                    n.nref.pref=new_node
                n.nref=new_node
    
    def add_before(self,data,x):
        if self.head==None:
            print("LL is empty")
        else:
            n=self.head
            while n is not None:
                if x==n.data:
                    break
                n=n.nref
            if n is None:
                print("x not in DLL")
            else:
                new_node=Node(data)
                new_node.nref=n
                new_node.pref=n.pref 
                if n.pref is not None :
                    n.pref.nref=new_node
                else:
                    self.head=new_node
                n.pref=new_node
    
    def del_begin(self):
        if self.head==None:
            print("DLL is empty")
            return
        if self.head.nref==None:
            self.head=None
            print("DLL is empty after deletion operation")
        else:
            self.head=self.head.nref
            self.head.pref=None
    
    def del_end(self):
        if self.head==None:
            print("DLL is empty")
            return
        if self.head.nref==None:
            self.head=None
            print("DLL is empty after deletion operation")
        else:
            n=self.head
            while n.nref is not None:
                n=n.nref
            n.pref.nref=None
            
    def del_by_value(self,x):
        if self.head==None:
            print("DLL is empty")
            return
        if self.head.nref==None:
            if x==self.head.data:
                self.head=None
            else:
                print("X is not in DDL")
        if self.head.data==x:
            self.head=self.head.nref
            self.head.pref=None
            return
        n=self.head
        while n.nref is not None:
            if x==n.data:
                break
            n=n.nref
        if n.nref is not None:
            n.nref.pref=n.pref
            n.pref.nref=n.nref
        else:
            if n.data==x:
                n.pref.nref=None
            else:
                print("x is not in DLL")
            
    
            
DLL=doubleLL()
DLL.add_begin(10)
DLL.add_begin(20)
DLL.add_begin(30)
DLL.add_begin(40)
DLL.add_end(-10)
DLL.add_end(-20)
DLL.add_after(100,30)
DLL.add_before(200,20)

DLL.del_begin()
DLL.del_by_value(100)
#DLL.backward_LL()
DLL.forward_LL()
```
